```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_market_data)
```

```python
# Create a DataFrame with the scaled data
df_scaled = pd.DataFrame(scaled_data, columns=df_market_data.columns)

# Copy the crypto names from the original DataFrame
df_scaled['coin_id'] = df_market_data.index

# Set the coin_id column as the index from the original DataFrame
df_scaled.set_index('coin_id', inplace=True)

# Display the scaled DataFrame
df_scaled.head()
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))
```

```python
# Create an empty list to store the inertia values
inertia_values = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list

# Looping through the range of k-values
for k in k_values:
    # 1. Create a KMeans model using the loop counter for the number of clusters
    kmeans = KMeans(n_clusters=k, random_state=42)
    
    # 2. Fit the model to the scaled data
    kmeans.fit(df_scaled)
    
    # 3. Append the model's inertia to the inertia list
    inertia_values.append(kmeans.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    "k": k_values,
    "inertia": inertia_values
}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.

elbow_curve_original = df_elbow.hvplot.line(
    x="k",
    y="inertia",
    title="Elbow Curve for KMeans Clustering",
    xlabel="Number of Clusters (k)",
    ylabel="Inertia",
    width=800,
    height=400,
    line_color="blue",
    label="Original Data"
)
elbow_curve_original
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** The best value for k is 4, with an inertia value of 79.022

---

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialize the K-Means model using the best value for k
optimal_k = 4  # Best value for k as from the elbow curve
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
```

```python
# Fit the K-Means model using the scaled DataFrame
kmeans.fit(df_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled DataFrame
cluster_labels = kmeans.predict(df_scaled)

# Print the resulting array of cluster values.
print(cluster_labels)
```

```python
# Create a copy of the scaled DataFrame
df_clustered = df_scaled.copy()
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters
df_clustered['Cluster'] = cluster_labels

# Display the copy of the scaled DataFrame
df_clustered.head()
```

```python
# Create a scatter plot using hvPlot by setting
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.

scatter_original = df_clustered.hvplot.scatter(
    x="price_change_percentage_24h",  # x-axis representing 24h price change percentage
    y="price_change_percentage_7d",   # y-axis representing 7d price change percentage
    c="Cluster",                     # Coloring the points by cluster label
    hover_cols=["coin_id"],           # Displaying the crypto name on hover
    title="Original Data Clusters",
    width=800,
    height=400,
    label="Original Data Clusters",
    cmap="Category10"
)
scatter_original
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce the original scaled DataFrame
# down to three principal components.
pca_data = pca.fit_transform(df_scaled)

# Creating a DataFrame to view the scaled PCA data
df_pca = pd.DataFrame(
    pca_data,                     # PCA components from the transformation
    columns=['PC1', 'PC2', 'PC3'],  # Column names for the three principal components
    index=df_scaled.index         # Retaining the original index (crypto names)
)

# Displaying the scaled PCA DataFrame
df_pca.head()
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component.
explained_variance = pca.explained_variance_ratio_

# Displaying the explained variance for each component
print("Explained Variance Ratio for each component:", explained_variance)
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** 0.89503166 
- 0.3719856 + 0.34700813 + 0.17603793 = 0.89503166
- This means that the three components together explain about 89.5% of the variance in the original crypto_market_data dataset.

```python
# Create a new DataFrame with the PCA data.
df_pca = pd.DataFrame(
    pca_data,                     # The PCA components from the transformation
    columns=['PC1', 'PC2', 'PC3'],  # Columns representing the 3 principal components
    index=df_scaled.index         # Retaining the original index (crypto names)
)

# Copy the crypto names from the original scaled DataFrame
df_pca['coin_id'] = df_scaled.index

# Set the coin_id column as index
df_pca.set_index('coin_id', inplace=True)

# Display the scaled PCA DataFrame
df_pca.head()
```

---

### Find the Best Value for k Using the Scaled PCA DataFrame

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))
```

```python
# Create an empty list to store the inertia values
inertia_values = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
# Create an empty list to store the inertia values

for k in k_values:
    # 1. Creating a KMeans model using the loop counter for the n_clusters
    kmeans = KMeans(n_clusters=k, random_state=42)
    
    # 2. Fitting the model to the data using `df_pca`
    kmeans.fit(df_pca)
    
    # 3. Appending the model.inertia_ to the inertia list
    inertia_values.append(kmeans.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    'k': k_values,              # k-values (number of clusters)
    'inertia': inertia_values   # Corresponding inertia values
}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
elbow_curve_pca = df_elbow.hvplot.line(
    x='k', 
    y='inertia', 
    title="Elbow Curve", 
    xlabel="Number of Clusters (k)", 
    ylabel="Inertia",
    line_color="red",
    label="PCA Data",
    width=800, 
    height=400,
)
elbow_curve_pca
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** The best value for k when using the PCA data is 4


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** The value differs from that of of the original data with regard to its inertia value of 49.665. This indicates a better definition of the clusters, as the clusters are more compact and have less dispersion.

### Cluster Cryptocurrencies with K-means Using the Scaled PCA DataFrame

```python
# Initialize the KMeans model with k=4 (best value from previous step)
kmeans_pca = KMeans(n_clusters=4, random_state=42)
```

```python
# Fit the K-Means model using the PCA data
kmeans_pca.fit(df_pca)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled PCA DataFrame
cluster_labels = kmeans_pca.predict(df_pca)

# Print the resulting array of cluster values
print(cluster_labels)
```

```python
# Create a copy of the scaled PCA DataFrame
df_pca_copy = df_pca.copy()

# Add a new column to the copy of the PCA DataFrame with the predicted clusters
df_pca_copy['Cluster'] = cluster_labels

# Display the copy of the scaled PCA DataFrame
df_pca_copy.head()
```

```python
# Create a scatter plot using hvPlot by setting
# `x="PC1"` and `y="PC2"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.

scatter_pca = df_pca_copy.hvplot.scatter(
    x='PC1', y='PC2',          # Plotting PC1 against PC2
    c='Cluster',               # Color by the 'Cluster' column
    hover_cols=['coin_id'],    # Display crypto names on hover
    title="Cryptocurrency Clusters in PCA Space (k=4)",
    xlabel="Principal Component 1 (PC1)",
    ylabel="Principal Component 2 (PC2)",
    width=800, 
    height=400,
    label="PCA Data Clusters",
    cmap="Category20"
)
scatter_pca
```
